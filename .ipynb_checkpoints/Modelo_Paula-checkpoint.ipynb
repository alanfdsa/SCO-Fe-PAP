# Função g.

Primeiro, tenho que obter a distribuição de spin similar ao modelo do Mario. Mas no modelo da Paula o parâmetro g depende da função partição magnética e de fônons. Então primeiro vou equacionar as funções partições Z_mag e Z_phonons. No artigo da Paula a definição é:

$g = \frac{Z_{HS}^{mag}}{Z_{LS}^{mag}}\times\frac{Z_{HS}^{phon}}{Z_{LS}^{phon}}$

onde 

$Z^{mag} = \frac{\sinh[(\frac{2J+1}{2J})x]}{\sinh(\frac{x}{2J})}$ e $x = \frac{g_J \mu_B J B}{k_B T}$

```python
import numpy as np

# Função Partição Magnética
def Z_mag(T, g_J, J, B):
    k_B = 1.380649e-23      # J/K
    mu_B = 9.2740100783e-24 # J/T
    if J == 0:
        return np.ones_like(T)
    x = (g_J * mu_B * J * B) / (k_B * T)
    
    num = np.sinh((2 * J + 1) * x / (2 * J))
    den = np.sinh(x / (2 * J))
    
    Z = num / den
    return Z

# Função Partição dos Fônons
def Z_phon(limite_soma, T, frequencias_cm):
    k_B = 1.380649e-23   # J/K
    hbar = 1.05457168e-34  # J.s
    c = 1
    beta = 1 / (k_B * T)
    
    if len(frequencias_cm) != limite_soma:
        raise ValueError(f"O número de frequências ({len(frequencias_cm)}) não bate com o limite da soma ({limite_soma}).")
    
    prod = np.ones_like(T, dtype=float)
    
    for omega_cm in frequencias_cm:
        omega = 2 * np.pi * c * omega_cm  # Convertendo de cm^-1 para rad/s
        x = -beta * hbar * omega
        num = np.exp(x / 2)
        den = 1 - np.exp(x)
        prod *= num / den

    return prod

# Teste
Temp = np.linspace(5, 500, 1000)
frequencias_HS = [20, 545.3, 965.2]
frequencias_LS = [74, 545.6, 968.5]
limite_soma = 3

Z_mag_HS = Z_mag(Temp, 2, 2, 0.01)
Z_mag_LS = Z_mag(Temp, 2, 0, 0.01)

Z_phon_HS = Z_phon(limite_soma, Temp, frequencias_HS)
Z_phon_LS = Z_phon(limite_soma, Temp, frequencias_LS)

g = (Z_mag_HS/Z_mag_LS)*(Z_phon_HS/Z_phon_LS)
```

```python
plt.plot(Temp,g)
```

```python
# Construindo e plotando a função partição de fônons para as frequências [10, 100, 500] cm^-1
import numpy as np
import matplotlib.pyplot as plt

# Constantes
k_B = 1.380649e-23       # J/K
hbar = 1.054571817e-34   # J.s
c_cm = 1     # cm/s

def Z_phon(T, frequencias_cm):
    """
    Z_phon(T) = produto sobre modos [ exp(-beta*hbar*omega/2) / (1 - exp(-beta*hbar*omega)) ]
    onde omega = 2*pi*c*(frequencia em cm^-1)
    T pode ser array.
    """
    T = np.asarray(T, dtype=float)
    beta = 1.0 / (k_B * T)
    Z = np.ones_like(T, dtype=float)
    for nu_cm in frequencias_cm:
        omega = 2 * np.pi * c_cm * nu_cm  # rad/s
        x = beta * hbar * omega          # = (ħω)/(k_B T)
        # para estabilidade numérica, computamos termo cuidadosamente
        # termo = exp(-x/2) / (1 - exp(-x)) = 1 / (2 sinh(x/2))
        # mas usaremos a forma segura usando np.expm1 quando apropriado
        with np.errstate(over=False, under=False, divide=False, invalid=False):
            exp_neg_x = np.exp(-x)
            denom = 1.0 - exp_neg_x
            term = np.exp(-x/2) / denom
            # para x muito pequeno (alto T), usamos limite assintótico: term ~ 1/x
            small_mask = x < 1e-6
            if np.any(small_mask):
                term[small_mask] = 1.0 / x[small_mask]  # limite: exp(-x/2)/(1-exp(-x)) ~ 1/x
        Z *= term
    return Z

# Parâmetros
frequencias = [10, 100, 500]  # cm^-1
T = np.linspace(5, 500, 500)   # K

Z = Z_phon(T, frequencias)

# Plot
plt.figure(figsize=(7,4))
plt.plot(T, Z, lw=2)
plt.xlabel('Temperatura (K)')
plt.ylabel('Função partição de fônons $Z_{\\rm phon}$')
plt.title('Z_phon vs T para frequências 10, 100, 500 cm$^{-1}$')
plt.yscale('log')  # escala log para melhor visualização, pois Z cresce muito
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
