# Função g.

Primeiro, tenho que obter a distribuição de spin similar ao modelo do Mario. Mas no modelo da Paula o parâmetro g depende da função partição magnética e de fônons. Então primeiro vou equacionar as funções partições Z_mag e Z_phonons. No artigo da Paula a definição é:

$g = \frac{Z_{HS}^{mag}}{Z_{LS}^{mag}}\times\frac{Z_{HS}^{phon}}{Z_{LS}^{phon}}$

onde 

$Z^{mag} = \frac{\sinh[(\frac{2J+1}{2J})x]}{\sinh(\frac{x}{2J})}$ e $x = \frac{g_J \mu_B J B}{k_B T}$

```python
import numpy as np

# Função Partição Magnética
def Z_mag(T, g_J, J, B):
    k_B = 1.380649e-23      # J/K
    mu_B = 9.2740100783e-24 # J/T
    if J == 0:
        return np.ones_like(T)
    x = (g_J * mu_B * J * B) / (k_B * T)
    
    num = np.sinh((2 * J + 1) * x / (2 * J))
    den = np.sinh(x / (2 * J))
    
    Z = num / den
    return Z

# Função Partição dos Fônons
def Z_phon(limite_soma, T, frequencias_cm):
    k_B = 1.380649e-23   # J/K
    hbar = 1.05457168e-34  # J.s
    c = 1
    beta = 1 / (k_B * T)
    
    if len(frequencias_cm) != limite_soma:
        raise ValueError(f"O número de frequências ({len(frequencias_cm)}) não bate com o limite da soma ({limite_soma}).")
    
    prod = np.ones_like(T, dtype=float)
    
    for omega_cm in frequencias_cm:
        omega = 2 * np.pi * c * omega_cm  # Convertendo de cm^-1 para rad/s
        x = -beta * hbar * omega
        num = np.exp(x / 2)
        den = 1 - np.exp(x)
        prod *= num / den

    return prod

# Teste
Temp = np.linspace(5, 500, 1000)
#frequencias_HS = [20, 545.3, 965.2]
#frequencias_LS = [74, 545.6, 968.5]
#limite_soma = 3

frequencias_HS = np.full(15,365.31*0.69503476)
frequencias_LS = np.full(15,190*0.69503476)
limite_soma = len(frequencias_HS)

Z_mag_HS = Z_mag(Temp, 2, 2, 0.01)
Z_mag_LS = Z_mag(Temp, 2, 0, 0.01)

Z_phon_HS = Z_phon(limite_soma, Temp, frequencias_HS)
Z_phon_LS = Z_phon(limite_soma, Temp, frequencias_LS)

g = (Z_mag_HS/Z_mag_LS)*(Z_phon_HS/Z_phon_LS)
```

```python
#!/usr/bin/env python3
"""
Versão corrigida do solver autoconsistente m(T) com:
 - termo não-linear (2 J1 m^3) / (K * V_LS) usando K em kBar e V_LS em Å^3 (V_LS = 2400 Å^3)
 - termo P·ΔV tratado como -0.5 * P * ΔV / 2 (equivalente a -0.25 * P·ΔV) com P em kBar e ΔV = 28 Å^3

Todas as conversões de unidades são feitas de forma explícita:
 1 kBar = 1e8 Pa
 1 Å^3 = 1e-30 m^3
 energias (J) -> temperaturas (K) dividindo por k_B

O restante do código (partições, ln g(T), Brillouin, etc.) foi mantido, apenas adaptando
os fatores e conversões para que a equação autoconsistente seja:

 m(T) = tanh( [ 2 J0_K * m + (2 J1_K * m^3) / (K * V_LS)_K
                 + 0.5 * T * ln g(T) - Δ_K - 0.25 * (P * ΔV)_K ] / T ).

Execute com: python3 solve_m_with_M_vs_T_corrected.py
"""
import numpy as np
import matplotlib.pyplot as plt

# constantes
k_B = 1.380649e-23  # J/K
mu_B = 9.2740100783e-24  # J/T (Bohr magneton)
KBAR_TO_PA = 1e8
A3_TO_M3 = 1e-30

# modelo: valores fornecidos
J0_K = 10.0
J1_K = 5.0
K_kBar = 0.0215          # K (elastic) in kBar as provided
V_LS_A3 = 2400.0         # V_LS in Å^3 (novo solicitado)
DeltaV_A3 = 28.0         # ΔV used in P·ΔV term (Å^3)

# partições e campos
S_HS = 2
S_LS = 0
omega_HS_K = 190.0
omega_LS_K = 365.31
B_field = 5.0          # campo usado no cálculo de M (T)
g_factor_HS = 2.0
g_factor_LS = 2.0

# novo Δ em Kelvin pedido
Delta_K_user = 750.0  # Δ = 750 K
# pressão padrão (kBar)
P_kBar_default = 0.5  # default P = 0

# ---- funções numéricas estáveis para ln Z ----
def log_sinh(z):
    z = np.array(z, dtype=float)
    z_abs = np.abs(z)
    res = np.empty_like(z_abs)
    small = z_abs < 1e-6
    large = z_abs > 50.0
    mid = ~(small | large)
    if np.any(small):
        za = z_abs[small]
        za_safe = np.where(za == 0.0, 1e-300, za)
        res[small] = np.log(za_safe) + (za ** 2) / 6.0
    if np.any(large):
        res[large] = z_abs[large] - np.log(2.0)
    if np.any(mid):
        res[mid] = np.log(np.sinh(z_abs[mid]))
    return res

def ln_Zmag(S, g_factor, B, T):
    T = np.array(T, dtype=float)
    if S == 0:
        return np.zeros_like(T)
    T_safe = np.where(T <= 0.0, 1e-300, T)
    y = (mu_B * B * g_factor) / (2.0 * k_B * T_safe)
    y = np.clip(y, -1e6, 1e6)
    A = (2.0 * S + 1.0) / (2.0 * S)
    Bcoef = 1.0 / (2.0 * S)
    ln_num = log_sinh(A * y)
    ln_den = log_sinh(Bcoef * y)
    lnZ = ln_num - ln_den
    return lnZ

def ln_Zphon(omega_K, T):
    T = np.array(T, dtype=float)
    T_safe = np.where(T <= 0.0, 1e-300, T)
    x = omega_K / T_safe
    x = np.clip(x, 0.0, 1e6)
    ln1mexp = np.empty_like(x)
    small = x < 1e-6
    large = x > 50.0
    mid = ~(small | large)
    if np.any(small):
        xs = x[small]
        ln1mexp[small] = np.log(np.where(xs == 0.0, 1e-300, xs)) - xs / 2.0
    if np.any(large):
        ln1mexp[large] = 0.0
    if np.any(mid):
        xm = x[mid]
        val = 1.0 - np.exp(-xm)
        val = np.where(val <= 0.0, 1e-300, val)
        ln1mexp[mid] = np.log(val)
    lnZ = 15.0 * (-0.5 * x - ln1mexp)
    return lnZ

def compute_ln_g(T_array, B=B_field, g_HS=g_factor_HS, g_LS=g_factor_LS):
    T = np.array(T_array, dtype=float)
    lnZmag_HS = ln_Zmag(S_HS, g_HS, B, T)
    lnZmag_LS = ln_Zmag(S_LS, g_LS, B, T)
    lnZphon_HS = ln_Zphon(omega_HS_K, T)
    lnZphon_LS = ln_Zphon(omega_LS_K, T)
    ln_g = (lnZmag_HS - lnZmag_LS) + (lnZphon_HS - lnZphon_LS)
    ln_g = np.where(np.isfinite(ln_g), ln_g, np.sign(ln_g) * 1e300)
    return ln_g, lnZmag_HS, lnZmag_LS, lnZphon_HS, lnZphon_LS

# ---- conversões utilitárias (para obter "K" em unidades de Kelvin) ----
def convert_K_times_V_to_K(K_kBar, V_A3):
    """
    Converte K (kBar) * V (Å^3) -> energia em Kelvin (energia/k_B)
    energy_J = K_Pa * V_m3
    energy_K = energy_J / k_B
    """
    K_Pa = K_kBar * KBAR_TO_PA
    V_m3 = V_A3 * A3_TO_M3
    energy_J = K_Pa * V_m3
    energy_K = energy_J / k_B
    return energy_K

def convert_PDV_to_K(P_kBar, DeltaV_A3):
    """
    Converte P (kBar) * DeltaV (Å^3) -> energia em Kelvin
    """
    P_Pa = P_kBar * KBAR_TO_PA
    DeltaV_m3 = DeltaV_A3 * A3_TO_M3
    energy_J = P_Pa * DeltaV_m3
    energy_K = energy_J / k_B
    return energy_K

# ---- solver autoconsistente corrigido ----
def solve_m_at_T(T, ln_g_T, J0_K=J0_K, J1_K=J1_K,
                 K_kBar=K_kBar, V_LS_A3=V_LS_A3,
                 Delta_K=Delta_K_user, P_kBar=0.0, DeltaV_A3=DeltaV_A3,
                 tol=1e-9, maxiter=2000, damping=0.5):
    """
    Resolve m = tanh( ARG / T ) com ARG em unidades Kelvin:
      ARG = 2 J0 m + (2 J1 m^3) / [K * V_LS]_K + 0.5 T ln g(T) - Delta_K - 0.25 * (P * DeltaV)_K

    Observações sobre o termo de pressão:
      - o usuário pediu que o termo -0.5 * PDV_K seja substituído por -0.5 * P * DeltaV / 2,
        isto é matematicamente -0.25 * P * DeltaV.
      - portanto, convertimos P*DeltaV para Kelvin e multiplicamos por 0.25 (com sinal negativo).
    """
    # denominação convertida (K * V_LS) em Kelvin
    #denom_K = convert_K_times_V_to_K(K_kBar, V_LS_A3)  # [K * V_LS] -> Kelvin
    denom_K = K_kBar*V_LS_A3*10**(-22)/k_B
    # P * DeltaV em Kelvin
    #PDV_K = convert_PDV_to_K(P_kBar, DeltaV_A3)
    PDV_K = P_kBar_default*DeltaV_A3*10**(-22)/k_B
    def rhs(m, T_local, ln_g_local):
        # evitar divisão por 0 em denom
        denom_safe = denom_K if denom_K != 0.0 else 1e-300
        # termo PDV conforme especificado: -0.25 * (P * DeltaV)_K
        PDV_term = 0.25 * PDV_K
        arg = (2.0 * J0_K * m
               + (2.0 * J1_K * (m ** 3)) / denom_safe
               + 0.5 * T_local * ln_g_local
               - Delta_K
               - PDV_term)
        return np.tanh(arg / T_local)

    initial_guesses = [0.0, 0.1, -0.1, 0.3, -0.3, 0.6, -0.6, 0.9, -0.9]
    solutions = []
    for m0 in initial_guesses:
        m = m0
        for it in range(maxiter):
            m_new = (1.0 - damping) * m + damping * rhs(m, T, ln_g_T)
            if np.isnan(m_new):
                break
            if abs(m_new - m) < tol:
                m = m_new
                break
            m = m_new
        if not np.isnan(m):
            resid = m - rhs(m, T, ln_g_T)
            if abs(resid) < 1e-6:
                solutions.append(m)
    if len(solutions) == 0:
        return np.nan
    sols = np.array(solutions)
    # eliminar duplicatas
    uniq = []
    for s in sols:
        if not any(abs(s - u) < 1e-6 for u in uniq):
            uniq.append(s)
    uniq = np.array(uniq)
    idx = np.argmax(np.abs(uniq))
    m_sol = float(uniq[idx])
    return m_sol

def solve_m_vs_T(T_array, Delta_K=Delta_K_user, P_kBar=0.0,
                 J0_K=J0_K, J1_K=J1_K, K_kBar=K_kBar,
                 V_LS_A3=V_LS_A3, DeltaV_A3=DeltaV_A3):
    T = np.array(T_array, dtype=float)
    ln_g_vals, *_ = compute_ln_g(T)
    m_vals = np.empty_like(T)
    # precompute converted constants for info
    denom_K_value = convert_K_times_V_to_K(K_kBar, V_LS_A3)
    PDV_K_value = convert_PDV_to_K(P_kBar, DeltaV_A3)
    for i, (Ti, ln_gi) in enumerate(zip(T, ln_g_vals)):
        m_vals[i] = solve_m_at_T(Ti, ln_gi, J0_K=J0_K, J1_K=J1_K,
                                  K_kBar=K_kBar, V_LS_A3=V_LS_A3,
                                  Delta_K=Delta_K, P_kBar=P_kBar, DeltaV_A3=DeltaV_A3)
    return m_vals, ln_g_vals, denom_K_value, PDV_K_value

# ---- Brillouin numericamente estável ----
def brillouin_BJ(J, x):
    J = float(J)
    x = np.array(x, dtype=float)
    if J == 0.0:
        return np.zeros_like(x)
    small = np.abs(x) < 1e-6
    big = np.abs(x) > 50.0
    mid = ~(small | big)
    BJ = np.empty_like(x)
    if np.any(small):
        BJ[small] = ((J + 1.0) / (3.0 * J)) * x[small]
    if np.any(big):
        BJ[big] = np.sign(x[big]) * 1.0
    if np.any(mid):
        xm = x[mid]
        a = (2.0 * J + 1.0) / (2.0 * J)
        b = 1.0 / (2.0 * J)
        coth_a = np.cosh(a * xm) / np.sinh(a * xm)
        coth_b = np.cosh(b * xm) / np.sinh(b * xm)
        BJ[mid] = a * coth_a - b * coth_b
    return BJ

# ---- execução principal: resolvemos m(T), calculamos N_HS e M(T) ----
if __name__ == "__main__":
    T_vals = np.linspace(1.0, 600.0, 600)  # 1 K ... 600 K
    # garantir P = 0 como pedido (pode alterar)
    P_use = 0.0
    m_vals, ln_g_vals, denom_K_val, PDV_K_val = solve_m_vs_T(
        T_vals, Delta_K=Delta_K_user, P_kBar=P_use,
        J0_K=J0_K, J1_K=J1_K, K_kBar=K_kBar,
        V_LS_A3=V_LS_A3, DeltaV_A3=DeltaV_A3)

    N_HS_vals = 0.5 * (1.0 + m_vals)

    # calcular y_HS e y_LS (vetorial)
    y_HS = (g_factor_HS * mu_B * B_field * S_HS) / (2.0 * k_B * T_vals)
    y_LS = (g_factor_LS * mu_B * B_field * S_LS) / (2.0 * k_B * T_vals)  # S_LS = 0 -> zeros

    # Brillouin
    BJ_HS = brillouin_BJ(S_HS, y_HS)
    BJ_LS = brillouin_BJ(S_LS, y_LS)  # zero array

    # Magnetização por centro em unidades de μ_B (M/μ_B)
    M_muB = (g_factor_HS * S_HS * N_HS_vals * BJ_HS) + (g_factor_LS * S_LS * (1.0 - N_HS_vals) * BJ_LS)
    M_SI = M_muB * mu_B

    # imprimir info e checagens de unidade
    print(f"Usando Δ = {Delta_K_user} K, P = {P_use} kBar, B = {B_field} T")
    print(f"K = {K_kBar} kBar, V_LS = {V_LS_A3} Å^3 -> (K * V_LS)_K = {denom_K_val:.6e} K")
    print(f"P * ΔV (ΔV={DeltaV_A3} Å^3) em unidades Kelvin = {PDV_K_val:.6e} K")
    print(f"Termo PDV usado no ARG: -0.25 * P·ΔV -> -0.25 * {PDV_K_val:.6e} K = {-0.25*PDV_K_val:.6e} K")

    # Plots: m(T), N_HS(T), M(T)/μ_B
    fig, axes = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

    ax = axes[0]
    ax.plot(T_vals, m_vals, label='m(T)', color='C0', lw=2)
    ax.plot(T_vals, N_HS_vals, label='N_HS(T) = (1+m)/2', color='C2', lw=2)
    ax.set_ylabel("m and N_HS")
    ax.set_ylim(-1.05, 1.05)
    ax.legend()
    ax.grid(alpha=0.3)

    ax2 = axes[1]
    ax2.plot(T_vals, M_muB, label='M / μ_B (per centro)', color='C1', lw=2)
    ax2.set_xlabel("T (K)")
    ax2.set_ylabel('M / μ_B (dimensionless)')
    ax2.set_ylim(bottom=min(np.nanmin(M_muB) - 0.05, -0.05))
    ax2.set_xlim(0,250)
    ax2.legend()
    ax2.grid(alpha=0.3)

    plt.suptitle(f"M(T) em B={B_field} T, Δ={Delta_K_user} K (corrigido)")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# constantes
k_B = 1.380649e-23  # J/K
mu_B = 9.2740100783e-24  # J/T (Bohr magneton)
KBAR_TO_PA = 1e8
A3_TO_M3 = 1e-30

# modelo: valores fornecidos
J0_K = 10.0
J1_K = 5.0
K_kBar = 0.0215          # K (elastic) in kBar as provided
V_LS_A3 = 2400.0         # V_LS in Å^3 (novo solicitado)
DeltaV_A3 = 28.0         # ΔV used in P·ΔV term (Å^3)

# partições e campos
S_HS = 2
S_LS = 0
omega_HS_K = 190.0
omega_LS_K = 365.31
B_field = 0.1          # campo usado no cálculo de M (T)
g_factor_HS = 2.0
g_factor_LS = 2.0

Delta_K_user = 750.0  # Δ = 750 K
P_kBar_default = 0.5  # default P = 0

#omega_HS_K = np.full(15,190)
#omega_LS_K = np.full(15,365.31)

#T = np.arange(1,500)


############### Funções que calculam a função partição de magnetização, fônons e por fim o ln(g) ############### 
def log_sinh(z):
    res = np.log(np.sinh(z))
    return res

def ln_Zmag(S,g_factor,B,T):
    T = np.array(T,dtype=float)
    if S==0:
        return np.zeros_like(T) # Se S = 0 o lnZ vai ser um array de zeros.
    T_safe = np.where(T <= 0.0, 1e-300, T) # Se T for negativo, retornar uma temperatura muito pequena.
    y = (mu_B * B * g_factor) / (2.0 * k_B * T_safe)
    #y = np.clip(y, -1e6, 1e6) # Se os valores vorem maiores ou menos que estes, serão recortados e colocados na borda de 0 ou 1.
    A = (2.0 * S + 1.0) / (2.0 * S)
    B = 1.0 / (2.0 * S)
    ln_num = log_sinh(A * y)
    ln_den = log_sinh(B * y)
    lnZ = ln_num - ln_den
    return lnZ

def ln_Zphon(omega_K, T):
    omega_K = np.array(omega_K, dtype=float)
    T = np.array(T, dtype=float)
    T_safe = np.where(T <= 0.0, 1e-300, T)
    x = omega_K[:, None] / T_safe[None, :]  # shape (n_omega, n_T)
    #x = np.clip(x, 0.0, 1e6)

    ln1mexp = np.empty_like(x)
    small = x < 1e-6
    large = x > 50.0
    mid = ~(small | large)
    # Small x approximation
    if np.any(small):
        xs = x[small]
        ln1mexp[small] = np.log(np.where(xs == 0.0, 1e-300, xs)) - xs / 2.0
    # Large x approximation
    if np.any(large):
        ln1mexp[large] = 0.0
    # Mid x standard calculation
    if np.any(mid):
        xm = x[mid]
        val = 1.0 - np.exp(-xm)
        val = np.where(val <= 0.0, 1e-300, val)
        ln1mexp[mid] = np.log(val)

    lnZ = np.sum(-0.5 * x - ln1mexp, axis=0)  # soma sobre omegas
    return lnZ

def compute_ln_g(T_array, B=B_field, g_HS=g_factor_HS, g_LS=g_factor_LS):
    T = np.array(T_array, dtype=float)
    lnZmag_HS = ln_Zmag(S_HS, g_HS, B, T)
    lnZmag_LS = ln_Zmag(S_LS, g_LS, B, T)
    lnZphon_HS = ln_Zphon(omega_HS_K, T)
    lnZphon_LS = ln_Zphon(omega_LS_K, T)
    ln_g = (lnZmag_HS - lnZmag_LS) + (lnZphon_HS - lnZphon_LS)
    #ln_g = np.where(np.isfinite(ln_g), ln_g, np.sign(ln_g) * 1e300)
    return ln_g, lnZmag_HS, lnZmag_LS, lnZphon_HS, lnZphon_LS
############### ############### ---- solver autoconsistente corrigido ---- ############### ############### ###############
def solve_m_at_T(T, ln_g_T, J0_K=J0_K, J1_K=J1_K,
                 K_kBar=K_kBar, V_LS_A3=V_LS_A3,
                 Delta_K=Delta_K_user, P_kBar=0.0, DeltaV_A3=DeltaV_A3,
                 tol=1e-9, maxiter=2000, damping=0.5):
    """
    Resolve m = tanh( ARG / T ) com ARG em unidades Kelvin:
      ARG = 2 J0 m + (2 J1 m^3) / [K * V_LS]_K + 0.5 T ln g(T) - Delta_K - 0.25 * (P * DeltaV)_K

    Observações sobre o termo de pressão:
      - o usuário pediu que o termo -0.5 * PDV_K seja substituído por -0.5 * P * DeltaV / 2,
        isto é matematicamente -0.25 * P * DeltaV.
      - portanto, convertimos P*DeltaV para Kelvin e multiplicamos por 0.25 (com sinal negativo).
    """
    # denominação convertida (K * V_LS) em Kelvin
    #denom_K = convert_K_times_V_to_K(K_kBar, V_LS_A3)  # [K * V_LS] -> Kelvin
    denom_K = K_kBar*V_LS_A3*10**(-22)/k_B
    # P * DeltaV em Kelvin
    #PDV_K = convert_PDV_to_K(P_kBar, DeltaV_A3)
    PDV_K = P_kBar_default*DeltaV_A3*10**(-22)/k_B
    def rhs(m, T_local, ln_g_local):
        # evitar divisão por 0 em denom
        denom_safe = denom_K if denom_K != 0.0 else 1e-300
        # termo PDV conforme especificado: -0.25 * (P * DeltaV)_K
        PDV_term = 0.25 * PDV_K
        arg = (2.0 * J0_K * m
               + (2.0 * J1_K * (m ** 3)) / denom_safe
               + 0.5 * T_local * ln_g_local
               - Delta_K
               - PDV_term)
        return np.tanh(arg / T_local)

    initial_guesses = [0.0, 0.1, -0.1, 0.3, -0.3, 0.6, -0.6, 0.9, -0.9]
    solutions = []
    for m0 in initial_guesses:
        m = m0
        for it in range(maxiter):
            m_new = (1.0 - damping) * m + damping * rhs(m, T, ln_g_T)
            if np.isnan(m_new):
                break
            if abs(m_new - m) < tol:
                m = m_new
                break
            m = m_new
        if not np.isnan(m):
            resid = m - rhs(m, T, ln_g_T)
            if abs(resid) < 1e-6:
                solutions.append(m)
    if len(solutions) == 0:
        return np.nan
    sols = np.array(solutions)
    # eliminar duplicatas
    uniq = []
    for s in sols:
        if not any(abs(s - u) < 1e-6 for u in uniq):
            uniq.append(s)
    uniq = np.array(uniq)
    idx = np.argmax(np.abs(uniq))
    m_sol = float(uniq[idx])
    return m_sol
```

```python
# Teste ln(g) = ln(Z^{mag}_{HS}/Z^{mag}_{LS}) = lnn(Z^{mag}_{HS}) - ln(Z^{mag}_{LS})

omega_HS_K = np.full(15,190)
omega_LS_K = np.full(15,365.31)

T = np.arange(1,500)

lnZ_HS_mag = ln_Zmag(2,2,0.1,T)
lnZ_LS_mag = ln_Zmag(0,2,0.1,T)

lnZ_HS_phon = ln_Zphon(omega_HS_K,T)
lnZ_LS_phon = ln_Zphon(omega_LS_K,T)

ln_g = compute_ln_g(T)

plt.plot(T, lnZ_HS_mag-lnZ_LS_mag,label='Diff Mag ln(HS-LS)')
plt.plot(T, lnZ_HS_phon-lnZ_LS_phon,label = 'Diff Phonon ln(HS-LS)')
plt.plot(T, lnZ_HS_mag-lnZ_LS_mag+lnZ_HS_phon-lnZ_LS_phon, label='ln(g) Manual')
plt.plot(T,ln_g[0], label='ln(g) Function')

plt.legend()
plt.show()
```
