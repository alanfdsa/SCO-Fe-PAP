A equação para entropia de configuração é:

#### $S_C(P,T) = -R \left \{ n_{HS} \ln(n_{hs}) + (1-n_{hs})\ln(1-n_{hs})\right \}$

onde o $n_{hs}$ é obtido no bloco anterior.

```python
R = 8.314 # Em J/molK

def S_C(n_hs):
    S = -R/M*(n_hs*np.log(n_hs) + (1-n_hs)*np.log(1-n_hs))
    return S

#S_C_amb = S_C((m_amb+1)/2)
```

# Entropia de Spin

Aqui teremos que:

$S_S(P,T,B) = n_{HS}S^{HS}_S(T,B) - (1-n_{HS})S^{LS}_S(T,B)$

onde

$S_S^i(T,B) = R (\ln{(Z)} - x_i m_i)$

$Z = \sinh{(a_i x_i)}/\sinh{(b_i x_i)}$

$a_i = 1+b_i$

$b_i = 1/2j_i$

$x_i = g_i J_i \beta \mu_B B$

Note que no artigo original é dado que $S_{HS} = 2$ e $S_{LS} = 0$ para o Ferro $d^6$. Assim ocorre o quench total do estado LS. Já no caso HS isso não é verdade porque tem quatro elétrons desemparelhados. E então tecnicamente teria uma contribuição pro L? De qualquer modo nos cálculos abaixo eu considerei que L = 0 e então somente o spin contribui. Em ambos os casos em que L = 0 temos então que J = S, logo $g_i = 2$ para ambos os casos. Como $J_{LS} = 0$ então o estado LS não vai contribuir. O valor do campo magnético eu coloquei qualquer um: testei 0.01, 0.1 e 1 T e não muda em nada o deltaS. Na última frase. Como no final tem que somar todas as contribuições, então apliquei a temperatura obtida via interpolação.

```python
def S_S(T, g_J, J, B):
    """
    Calcula a entropia magnética (em J/mol·K) de um paramagneto ideal de spin J.

    Entradas:
    - T : Temperatura (K)
    - g_J : Fator de Landé
    - J : Momento total de spin
    - B : Campo magnético (Tesla)

    Saída:
    - S : Entropia (J/mol·K)
    """
    k_B = 1.380649e-23      # J/K
    mu_B = 9.2740100783e-24 # J/T
    R = 8.314462618         # J/mol·K

    x = (g_J * mu_B * J * B) / (k_B * T)

    num = np.sinh((2 * J + 1) * x / (2 * J))
    den = np.sinh(x / (2 * J))
    Z = num / den

    coeff = (2 * J + 1) / (2 * J)
    B_J = (coeff / np.tanh(coeff * x)) - (1 / (2 * J * np.tanh(x / (2 * J))))

    S = R/M*(np.log(Z) + x * B_J)
    return S
```

```python
S_S_amb = n_hs_amb_interp*S_S(T_common, 2, 2,0.01)
S_S_2kbar = n_hs_2kbar_interp*S_S(T_common, 2, 2,0.01)

delta_S_Spin = -(S_S_2kbar - S_S_amb)

#plt.plot(T_common,S_S_amb/M,label='Pressão ambiente')
#plt.plot(T_common,S_S_2kbar/M,label='Pressão 2kbar')

plt.scatter(delta_S_p2_rev_comp['T(K)'],delta_S_p2_rev_comp['DS(J/kgK)'],color = 'red',label='Rev')
plt.plot(T_common, delta_S_Spin,label='$\Delta S_{spin}$')
plt.plot(T_common,delta_S_Config,label='$\Delta S_{config}$')
plt.plot(T_common,delta_S_Config+delta_S_Spin/M,label='$\Delta S_{soma}$=$\Delta S_{spin}$+$\Delta S_{config}$')

#plt.ylim(-3,10)
plt.xlim(300,330)
plt.ylabel('Entropia (J/kg$ \cdot $K)')
plt.xlabel('Temperatura (K)')

plt.legend()
plt.show()

```

# Entropia de vibração: Termos ópticos

Aqui temos:

## $S_O(T,P) = n_{HS}S_O^{HS} + (1-n_{HS})S_O^{LS}$

onde 

## $S^i_O (T) = R \sum_{\nu} \left \{-\ln{(1-e^{-\beta \hbar \omega^i_\nu})}   + \frac{\beta \hbar \omega^i_\nu}{e^{\beta \hbar \omega^i_\nu - 1}}         \right \}$

Aqui os valores de frequência que eu utilizei foram os valores médios dados por (1450,900 e 72) e (1450,900,264), que foram valores teste que coloquei. Preciso das frequências reais dessa molécula.

```python
import numpy as np
import matplotlib.pyplot as plt

def S_O(limite_soma, T, frequencias_cm):
    """
    Calcula a expressão da imagem com frequências em cm^-1.

    Parâmetros:
    limite_soma (int): Número de termos da soma.
    T (float): Temperatura em Kelvin.
    frequencias_cm (list ou array): Lista de frequências (em cm^-1).

    Retorno:
    float: Valor da expressão (J/mol·K).
    """
    k_B = 1.380649e-23       # J/K
    hbar = 1.054571817e-34   # J.s
    c = 1        # 
    R = 8.314462618          # J/mol·K

    beta = 1 / (k_B * T)
    soma = 0.0

    if len(frequencias_cm) != limite_soma:
        raise ValueError(f"O número de frequências ({len(frequencias_cm)}) não bate com o limite da soma ({limite_soma}).")

    for omega_cm in frequencias_cm:
        omega = 2 * np.pi * c * omega_cm  # Convertendo de cm^-1 para rad/s
        x = beta * hbar * omega

        if x == 0:
            continue  # Evita divisão por zero

        termo_log = -np.log(1 - np.exp(-x))
        termo_fracao = x / (np.exp(x) - 1)

        soma += termo_log + termo_fracao

    return R/M * soma

# Parâmetros para o gráfico

frequencias_HS = [20,545.3,965.2]
frequencias_LS = [74,545.6,968.5]

limite_soma = 3
temperaturas = T_common  # Temperaturas de 10K a 500K

# Calculando S para cada temperatura
#S_valores = [S_O(limite_soma, T, frequencias_cm) for T in temperaturas]
S_O_HS = [S_O(limite_soma, T, frequencias_HS) for T in temperaturas]
S_O_LS = [S_O(limite_soma, T, frequencias_LS) for T in temperaturas]

S_O_2kbar = n_hs_2kbar_interp*S_O_HS + (1-n_hs_2kbar_interp)*S_O_LS
S_O_amb =  n_hs_amb_interp*S_O_HS + (1-n_hs_amb_interp)*S_O_LS
delta_S_O = -(S_O_2kbar-S_O_amb)
plt.xlim(250,450)
plt.scatter(delta_S_p2_rev_comp['T(K)'],delta_S_p2_rev_comp['DS(J/kgK)'],color = 'black',label='Experimental')
plt.plot(T_common,delta_S_O,label='$\Delta S_{opt}$')
plt.plot(T_common, delta_S_Spin,label='$\Delta S_{spin}$')
plt.plot(T_common,delta_S_Config,label='$\Delta S_{config}$')
plt.plot(T_common,delta_S_Config+delta_S_Spin+delta_S_O,label='$\Delta S_{total}$=$\Delta S_{spin}$+$\Delta S_{config}+\Delta S_{opt}$')
plt.ylabel("$\Delta$S (J/Kg$\cdot$K)")
plt.xlabel("Temperature (K)")

#plt.legend()
plt.show()
```

# Otimização g

Aqui, eu deixo todos os parâmetros fixos, com exceção do g, que varia em um determinado range.
O efeito é que a medida que g aumenta, mais as curvas de entropia se deslocam pra esquerda. Ao mesmo tempo, também é observado que o valor máximo de deltaS aumenta, a medida que g aumenta (mas nada tão substancial). Por exemplo, variando g de 1000 até 4900, o delta_S_max aumentou em 100.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Criar diretório para salvar os gráficos
output_dir = "g_variation_plots"
os.makedirs(output_dir, exist_ok=True)

# A massa molar, extraída da foŕmula quimica que o artigo passa é:
M = 0.540273 #kg/mol

delta_S_p2_rev_comp = pd.read_csv("delta_S_p2_rev_comp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_rev_decomp = pd.read_csv("delta_S_p2_rev_decomp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_comp = pd.read_csv("delta_S_p2_comp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_decomp = pd.read_csv("delta_S_p2_decomp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})

########################################################## Parâmetros ##########################################################
# Parâmetros
T_0 = 300 # Em Kelvin
#T_0 = 262
Delta_0 = 2050 # Em Kelvin
delta_V = 48e-30 # A unidade dá em metros ao cubo.
k_B = 1.380649e-23 # Em J/K

# Loop sobre valores de g
for g in range(1000, 5000, 100):
    ########################################################## Distribuição de Spin ##########################################################
    def T(m, delta_0, delta_V, P, g, T0):
    #    P_linha = (((delta_V)**(1/3)) * P) / 4
        P_linha = (((delta_V) * P)) / 4
        #gamma = (k_B * delta_0 + P_linha * (delta_V**3)) / (k_B * T0)
        gamma = (delta_0 + P_linha/k_B) / (T0)
        numerator = 2 * (m - gamma)
        denominator = np.log((m + 1) / (g**2 * (1 - m)))
        return T0 * (numerator / denominator)

    # Varredura original em m
    m_amb = np.linspace(-0.999, 0.999, 10000)
    m_2kbar = np.linspace(-0.999, 0.999, 10000)

    Temp_amb = T(m_amb, Delta_0, delta_V, 101325, g, T_0)
    Temp_2kbar = T(m_2kbar, Delta_0, delta_V, 200e6, g, T_0)

    n_hs_amb = (m_amb + 1) / 2
    n_hs_2kbar = (m_2kbar + 1) / 2

    # Define um grid comum de temperatura (exemplo: de 1K a 300K)
    T_min = max(np.min(Temp_amb), np.min(Temp_2kbar))
    T_max = min(np.max(Temp_amb), np.max(Temp_2kbar))
    T_common = np.linspace(T_min, T_max, 1000)

    # Interpolação para obter n_hs em T_common
    n_hs_amb_interp = np.interp(T_common, Temp_amb, n_hs_amb)
    n_hs_2kbar_interp = np.interp(T_common, Temp_2kbar, n_hs_2kbar)

    m_amb = np.linspace(-0.999, 0.999, 10000)
    m_2kbar = np.linspace(-0.999, 0.999, 10000)

    Temp_amb = T(m_amb, Delta_0, delta_V, 101325, g, T_0)
    Temp_2kbar = T(m_2kbar, Delta_0, delta_V, 200e6, g, T_0)
    n_hs_amb = (m_amb + 1) / 2
    n_hs_2kbar = (m_2kbar + 1) / 2

    ########################################################## Entropia Configuracional ##########################################################
    R = 8.314 # Em J/molK

    def S_C(n_hs):
        S = -R/M*(n_hs*np.log(n_hs) + (1-n_hs)*np.log(1-n_hs))
        return S

    S_C_amb = S_C(n_hs_amb_interp)
    S_C_2kbar = S_C(n_hs_2kbar_interp)
    delta_S_Config = -(S_C_2kbar-S_C_amb)
    ########################################################## Entropia de Spin ##########################################################
    def S_S(T, g_J, J, B):
        """
        Calcula a entropia magnética (em J/mol·K) de um paramagneto ideal de spin J.

        Entradas:
        - T : Temperatura (K)
        - g_J : Fator de Landé
        - J : Momento total de spin
        - B : Campo magnético (Tesla)

        Saída:
        - S : Entropia (J/mol·K)
        """
        k_B = 1.380649e-23      # J/K
        mu_B = 9.2740100783e-24 # J/T
        R = 8.314462618         # J/mol·K

        x = (g_J * mu_B * J * B) / (k_B * T)

        num = np.sinh((2 * J + 1) * x / (2 * J))
        den = np.sinh(x / (2 * J))
        Z = num / den

        coeff = (2 * J + 1) / (2 * J)
        B_J = (coeff / np.tanh(coeff * x)) - (1 / (2 * J * np.tanh(x / (2 * J))))

        S = R /M* (np.log(Z) + x * B_J)
        return S

    S_S_amb = n_hs_amb_interp*S_S(T_common, 2, 2,0.01)
    S_S_2kbar = n_hs_2kbar_interp*S_S(T_common, 2, 2,0.01)

    delta_S_Spin = -(S_S_2kbar - S_S_amb)
    ########################################################## Entropia Óptica ##########################################################
    def S_O(limite_soma, T, frequencias_cm):
        """
        Calcula a expressão da imagem com frequências em cm^-1.

        Parâmetros:
        limite_soma (int): Número de termos da soma.
        T (float): Temperatura em Kelvin.
        frequencias_cm (list ou array): Lista de frequências (em cm^-1).

        Retorno:
        float: Valor da expressão (J/mol·K).
        """
        k_B = 1.380649e-23       # J/K
        hbar = 1.054571817e-34   # J.s
        c = 1        # 
        R = 8.314462618          # J/mol·K

        beta = 1 / (k_B * T)
        soma = 0.0

        if len(frequencias_cm) != limite_soma:
            raise ValueError(f"O número de frequências ({len(frequencias_cm)}) não bate com o limite da soma ({limite_soma}).")

        for omega_cm in frequencias_cm:
            omega = 2 * np.pi * c * omega_cm  # Convertendo de cm^-1 para rad/s
            x = beta * hbar * omega

            if x == 0:
                continue  # Evita divisão por zero

            termo_log = -np.log(1 - np.exp(-x))
            termo_fracao = x / (np.exp(x) - 1)

            soma += termo_log + termo_fracao

        return R * soma /M


    frequencias_HS = [20,545.3,965.2]
    frequencias_LS = [74,545.6,968.5]
    frequencias_HS = [20,545.3,965.2]
    frequencias_LS = [74,545.6,968.5]

    limite_soma = 3
    temperaturas = T_common  # Temperaturas de 10K a 500K

    # Calculando S para cada temperatura
    #S_valores = [S_O(limite_soma, T, frequencias_cm) for T in temperaturas]
    S_O_HS = [S_O(limite_soma, T, frequencias_HS) for T in temperaturas]
    S_O_LS = [S_O(limite_soma, T, frequencias_LS) for T in temperaturas]

    S_O_2kbar = n_hs_2kbar_interp*S_O_HS + (1-n_hs_2kbar_interp)*S_O_LS
    S_O_amb =  n_hs_amb_interp*S_O_HS + (1-n_hs_2kbar_interp)*S_O_LS
    delta_S_O = -(S_O_2kbar-S_O_amb)

    ########################################################## Plot dos Gráficos ##########################################################
    plt.figure(figsize=(10, 6))
    plt.xlim(250,450)
    plt.scatter(delta_S_p2_rev_comp['T(K)'],delta_S_p2_rev_comp['DS(J/kgK)'],color = 'black',label='Experimental')
    plt.plot(T_common,delta_S_O,label='$\Delta S_{opt}$')
    plt.plot(T_common, delta_S_Spin,label='$\Delta S_{spin}$')
    plt.plot(T_common,delta_S_Config,label='$\Delta S_{config}$')
    plt.plot(T_common,delta_S_Config+delta_S_Spin+delta_S_O,label='$\Delta S_{total}$=$\Delta S_{spin}$+$\Delta S_{config}+\Delta S_{opt}$')
    plt.ylabel("$\Delta$S (J/Kg$\cdot$K)")
    plt.xlabel("Temperature (K)")
    plt.title(f"g = {g}")
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Salvar o gráfico
    filename = os.path.join(output_dir, f"g_{g}.png")
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()

print("Processamento concluído! Gráficos salvos na pasta 'g_variation_plots'")
```

# Otimização g e delta_0

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.optimize import minimize

# Criar diretório para salvar os gráficos
output_dir = "g_Delta0_optimization_plots"
os.makedirs(output_dir, exist_ok=True)

# A massa molar, extraída da foŕmula quimica que o artigo passa é:
M = 0.540273 #kg/mol

delta_S_p2_rev_comp = pd.read_csv("delta_S_p2_rev_comp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_rev_decomp = pd.read_csv("delta_S_p2_rev_decomp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_comp = pd.read_csv("delta_S_p2_comp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})
delta_S_p2_decomp = pd.read_csv("delta_S_p2_decomp.csv",header = None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})

########################################################## Parâmetros ##########################################################
# Parâmetros fixos
T_0 = 313 # Em Kelvin fixo
delta_V = 120e-30 # A unidade dá em metros ao cubo.
k_B = 1.380649e-23 # Em J/K

# Função para calcular a função objetivo (quanto mais próximo de 0, melhor)
def objective_function(params):
    g, Delta_0 = params
    
    ########################################################## Distribuição de Spin ##########################################################
    def T(m, delta_0, delta_V, P, g, T0):
        P_linha = (((delta_V) * P)) / 4
        gamma = (delta_0 + P_linha/k_B) / (T0)
        numerator = 2 * (m - gamma)
        denominator = np.log((m + 1) / (g**2 * (1 - m)))
        return T0 * (numerator / denominator)

    # Varredura original em m
    m_amb = np.linspace(-0.999, 0.999, 10000)
    m_2kbar = np.linspace(-0.999, 0.999, 10000)

    Temp_amb = T(m_amb, Delta_0, delta_V, 101325, g, T_0)
    Temp_2kbar = T(m_2kbar, Delta_0, delta_V, 200e6, g, T_0)

    n_hs_amb = (m_amb + 1) / 2
    n_hs_2kbar = (m_2kbar + 1) / 2

    # Define um grid comum de temperatura
    T_min = max(np.min(Temp_amb), np.min(Temp_2kbar))
    T_max = min(np.max(Temp_amb), np.max(Temp_2kbar))
    T_common = np.linspace(T_min, T_max, 1000)

    # Interpolação para obter n_hs em T_common
    n_hs_amb_interp = np.interp(T_common, Temp_amb, n_hs_amb)
    n_hs_2kbar_interp = np.interp(T_common, Temp_2kbar, n_hs_2kbar)

    ########################################################## Entropia Configuracional ##########################################################
    R = 8.314 # Em J/molK

    def S_C(n_hs):
        S = -R/M*(n_hs*np.log(n_hs) + (1-n_hs)*np.log(1-n_hs))
        return S

    S_C_amb = S_C(n_hs_amb_interp)
    S_C_2kbar = S_C(n_hs_2kbar_interp)
    delta_S_Config = -(S_C_2kbar-S_C_amb)
    
    ########################################################## Entropia de Spin ##########################################################
    def S_S(T, g_J, J, B):
        k_B = 1.380649e-23      # J/K
        mu_B = 9.2740100783e-24 # J/T
        R = 8.314462618         # J/mol·K

        x = (g_J * mu_B * J * B) / (k_B * T)

        num = np.sinh((2 * J + 1) * x / (2 * J))
        den = np.sinh(x / (2 * J))
        Z = num / den

        coeff = (2 * J + 1) / (2 * J)
        B_J = (coeff / np.tanh(coeff * x)) - (1 / (2 * J * np.tanh(x / (2 * J))))

        S = R /M* (np.log(Z) + x * B_J)
        return S

    S_S_amb = n_hs_amb_interp*S_S(T_common, 2, 2,0.01)
    S_S_2kbar = n_hs_2kbar_interp*S_S(T_common, 2, 2,0.01)
    delta_S_Spin = -(S_S_2kbar - S_S_amb)
    
    ########################################################## Entropia Óptica ##########################################################
    def S_O(limite_soma, T, frequencias_cm):
        k_B = 1.380649e-23       # J/K
        hbar = 1.054571817e-34   # J.s
        c = 1        
        R = 8.314462618          # J/mol·K

        beta = 1 / (k_B * T)
        soma = 0.0

        for omega_cm in frequencias_cm:
            omega = 2 * np.pi * c * omega_cm
            x = beta * hbar * omega

            if x == 0:
                continue

            termo_log = -np.log(1 - np.exp(-x))
            termo_fracao = x / (np.exp(x) - 1)
            soma += termo_log + termo_fracao

        return R /M * soma

    frequencias_HS = [20,545.3,965.2]
    frequencias_LS = [74,545.6,968.5]
    limite_soma = 3

    S_O_HS = [S_O(limite_soma, T, frequencias_HS) for T in T_common]
    S_O_LS = [S_O(limite_soma, T, frequencias_LS) for T in T_common]

    S_O_2kbar = n_hs_2kbar_interp*S_O_HS + (1-n_hs_2kbar_interp)*S_O_LS
    S_O_amb =  n_hs_amb_interp*S_O_HS + (1-n_hs_amb_interp)*S_O_LS
    delta_S_O = -(S_O_2kbar-S_O_amb)
    
    ########################################################## Cálculo da função objetivo ##########################################################
    delta_S_total = delta_S_Config + delta_S_Spin + delta_S_O
    
    # Encontrar máximo e temperatura correspondente
    max_delta_S = np.max(delta_S_total)
    max_temp = T_common[np.argmax(delta_S_total)]
    
    # Função objetivo: penaliza desvios do máximo (80-100) e temperatura (313K)
    target_max = 70  # Valor médio da faixa 80-100
    target_temp = 313
    
    error_max = abs(max_delta_S - target_max) / target_max
    error_temp = abs(max_temp - target_temp) / target_temp
    
    # Peso maior para a temperatura (mais importante)
    return error_temp * 0.7 + error_max * 0.3

# Otimização para encontrar os melhores parâmetros
initial_guess = [5000, 1000]  # Valores iniciais
bounds = [(100, 10000), (500, 50000)]  # Limites para g e Delta_0

result = minimize(objective_function, initial_guess, bounds=bounds, method='L-BFGS-B')
optimal_g, optimal_Delta_0 = result.x

print(f"Parâmetros ótimos encontrados:")
print(f"g = {optimal_g:.2f}")
print(f"Delta_0 = {optimal_Delta_0:.2f} K")
print(f"Valor da função objetivo: {result.fun:.6f}")

# Agora gerar o gráfico com os parâmetros ótimos
g = optimal_g
Delta_0 = optimal_Delta_0

########################################################## Cálculo completo com parâmetros ótimos ##########################################################
def T(m, delta_0, delta_V, P, g, T0):
    P_linha = (((delta_V) * P)) / 4
    gamma = (delta_0 + P_linha/k_B) / (T0)
    numerator = 2 * (m - gamma)
    denominator = np.log((m + 1) / (g**2 * (1 - m)))
    return T0 * (numerator / denominator)

m_amb = np.linspace(-0.999, 0.999, 10000)
m_2kbar = np.linspace(-0.999, 0.999, 10000)

Temp_amb = T(m_amb, Delta_0, delta_V, 101325, g, T_0)
Temp_2kbar = T(m_2kbar, Delta_0, delta_V, 200e6, g, T_0)

n_hs_amb = (m_amb + 1) / 2
n_hs_2kbar = (m_2kbar + 1) / 2

T_min = max(np.min(Temp_amb), np.min(Temp_2kbar))
T_max = min(np.max(Temp_amb), np.max(Temp_2kbar))
T_common = np.linspace(T_min, T_max, 1000)

n_hs_amb_interp = np.interp(T_common, Temp_amb, n_hs_amb)
n_hs_2kbar_interp = np.interp(T_common, Temp_2kbar, n_hs_2kbar)

R = 8.314
def S_C(n_hs):
    return -R/M*(n_hs*np.log(n_hs) + (1-n_hs)*np.log(1-n_hs))

S_C_amb = S_C(n_hs_amb_interp)
S_C_2kbar = S_C(n_hs_2kbar_interp)
delta_S_Config = -(S_C_2kbar-S_C_amb)

def S_S(T, g_J, J, B):
    k_B = 1.380649e-23
    mu_B = 9.2740100783e-24
    R = 8.314462618

    x = (g_J * mu_B * J * B) / (k_B * T)
    num = np.sinh((2 * J + 1) * x / (2 * J))
    den = np.sinh(x / (2 * J))
    Z = num / den

    coeff = (2 * J + 1) / (2 * J)
    B_J = (coeff / np.tanh(coeff * x)) - (1 / (2 * J * np.tanh(x / (2 * J))))

    return R /M* (np.log(Z) + x * B_J)

S_S_amb = n_hs_amb_interp*S_S(T_common, 2, 2,0.01)
S_S_2kbar = n_hs_2kbar_interp*S_S(T_common, 2, 2,0.01)
delta_S_Spin = -(S_S_2kbar - S_S_amb)

def S_O(limite_soma, T, frequencias_cm):
    k_B = 1.380649e-23
    hbar = 1.054571817e-34
    c = 1
    R = 8.314462618

    beta = 1 / (k_B * T)
    soma = 0.0

    for omega_cm in frequencias_cm:
        omega = 2 * np.pi * c * omega_cm
        x = beta * hbar * omega
        if x == 0:
            continue
        termo_log = -np.log(1 - np.exp(-x))
        termo_fracao = x / (np.exp(x) - 1)
        soma += termo_log + termo_fracao

    return R /M *soma

frequencias_HS = [20,545.3,965.2]
frequencias_LS = [74,545.6,968.5]
S_O_HS = [S_O(3, T, frequencias_HS) for T in T_common]
S_O_LS = [S_O(3, T, frequencias_LS) for T in T_common]

S_O_2kbar = n_hs_2kbar_interp*S_O_HS + (1-n_hs_2kbar_interp)*S_O_LS
S_O_amb =  n_hs_amb_interp*S_O_HS + (1-n_hs_amb_interp)*S_O_LS
delta_S_O = -(S_O_2kbar-S_O_amb)

delta_S_total = delta_S_Config + delta_S_Spin + delta_S_O
max_delta_S = np.max(delta_S_total)
max_temp = T_common[np.argmax(delta_S_total)]

print(f"\nResultados com parâmetros ótimos:")
print(f"Máximo de ΔS_total: {max_delta_S:.2f} J/kg·K")
print(f"Temperatura do máximo: {max_temp:.2f} K")

########################################################## Plot do Gráfico ##########################################################
plt.figure(figsize=(12, 8))
plt.xlim(250,450)
plt.scatter(delta_S_p2_rev_comp['T(K)'],delta_S_p2_rev_comp['DS(J/kgK)'],color = 'black',label='Experimental')
plt.plot(T_common,delta_S_O,label='$\Delta S_{opt}$')
plt.plot(T_common, delta_S_Spin,label='$\Delta S_{spin}$')
plt.plot(T_common,delta_S_Config,label='$\Delta S_{config}$')
plt.plot(T_common,delta_S_total,label='$\Delta S_{total}$=$\Delta S_{spin}$+$\Delta S_{config}+\Delta S_{opt}$', linewidth=3)

# Marcar o ponto máximo
plt.axvline(x=max_temp, color='red', linestyle='--', alpha=0.7, label=f'T_max = {max_temp:.1f}K')
plt.axhline(y=max_delta_S, color='red', linestyle='--', alpha=0.7, label=f'ΔS_max = {max_delta_S:.1f} J/kg·K')

plt.ylabel("$\Delta$S (J/Kg$\cdot$K)")
plt.xlabel("Temperature (K)")
plt.title(f"Parâmetros ótimos: g = {g:.1f}, Δ₀ = {Delta_0:.1f} K")
plt.legend()
plt.grid(True, alpha=0.3)

# Salvar o gráfico
filename = os.path.join(output_dir, f"optimal_g_{g:.1f}_Delta0_{Delta_0:.1f}.png")
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()

print(f"Gráfico salvo: {filename}")
```

# Otimização completa


```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from itertools import product

# ================================
# CONFIGURAÇÃO: controle dos parâmetros
# modes: "optimize" | "fixed" | "sweep"
# ================================
param_control = {
    "g": {
        "mode": "fixed",
        "value": 4961.41,            # usado apenas se mode="fixed" ou como valor-base
        "bounds": (100.0, 10000.0)  # obrigatório se mode="optimize"
    },
    "Delta_0": {
        "mode": "fixed",
        "value": 2623.55,
        "bounds": (500.0, 50000.0)
    },
    "T_0": {
        "mode": "optimize",
        "value": 313.0,
        "bounds": (100.0, 400.0)
    },
    "delta_V": {
        "mode": "optimize",                     # <<< varrer ΔV conforme seu pedido
        "value": 43e-30,                     # usado se "fixed"
        "bounds": (10e-30, 100e-30),         # usado se "optimize"
        # valores para varrer; você pode usar np.arange aqui:
        "values": list(np.arange(10e-30, 100e-30, 5e-30))
    }
}

# ================================
# Outras configurações
# ================================
N_TESTS = 5  # número de testes por combinação de sweep
OUTPUT_DIR = "optimization_plots_param_control"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ================================
# Dados experimentais
# ================================
M = 0.540273  # kg/mol
k_B = 1.380649e-23  # J/K

# carregue somente o dataset que você usa no scatter
delta_S_p2_rev_comp = pd.read_csv("delta_S_p2_rev_comp.csv", header=None).rename(columns={0:"T(K)",1:"DS(J/kgK)"})


# ================================
# Modelo físico: funções auxiliares
# ================================
def T_of_m(m, delta_0, delta_V, P, g, T0):
    # Equações conforme seu código
    P_linha = (delta_V * P) / 4.0
    gamma = (delta_0 + P_linha / k_B) / T0
    numerator = 2.0 * (m - gamma)
    denom_arg = (m + 1.0) / (g**2 * (1.0 - m))
    # Evitar problemas numéricos
    denom_arg = np.clip(denom_arg, 1e-300, None)
    denominator = np.log(denom_arg)
    return T0 * (numerator / denominator)

def S_C_from_nhs(n_hs):
    # Entropia configuracional (J/kg·K)
    R = 8.314
    n = np.clip(n_hs, 1e-12, 1-1e-12)
    return -R/M * (n*np.log(n) + (1-n)*np.log(1-n))

def S_S(T, g_J, J, B):
    # Entropia de spin (J/kg·K) – seu código
    mu_B = 9.2740100783e-24 # J/T
    R = 8.314462618         # J/mol·K
    T = np.array(T)
    x = (g_J * mu_B * J * B) / (k_B * T)
    num = np.sinh((2 * J + 1) * x / (2 * J))
    den = np.sinh(x / (2 * J))
    den = np.clip(den, 1e-300, None)
    Z = num / den
    coeff = (2 * J + 1) / (2 * J)
    # Evitar 0 em tanh
    t1 = np.tanh(np.clip(coeff * x, -50, 50))
    t2 = np.tanh(np.clip(x / (2 * J), -50, 50))
    t1 = np.where(np.abs(t1) < 1e-300, np.sign(t1)*1e-300, t1)
    t2 = np.where(np.abs(t2) < 1e-300, np.sign(t2)*1e-300, t2)
    B_J = (coeff / t1) - (1.0 / (2 * J * t2))
    return R/M * (np.log(np.clip(Z, 1e-300, None)) + x * B_J)

def S_O(limite_soma, T, frequencias_cm):
    # Entropia "óptica" (J/kg·K) – conforme seu código
    hbar = 1.054571817e-34
    c = 1.0
    R = 8.314462618
    beta = 1.0 / (k_B * T)
    soma = 0.0
    for omega_cm in frequencias_cm:
        omega = 2.0 * np.pi * c * omega_cm
        x = beta * hbar * omega
        # Evitar over/underflow desnecessário
        x = np.clip(x, 1e-12, 1e3)
        soma += (-np.log1p(-np.exp(-x))) + x/(np.exp(x)-1.0)
    return R/M * soma

FREQ_HS = [20.0, 545.3, 965.2]
FREQ_LS = [74.0, 545.6, 968.5]

def compute_curves(g, Delta_0, T_0, delta_V):
    # Malhas em m e temperaturas p/ 1 atm e 2 kbar
    m_amb = np.linspace(-0.999, 0.999, 10000)
    m_2k  = np.linspace(-0.999, 0.999, 10000)

    Temp_amb = T_of_m(m_amb, Delta_0, delta_V, 101325.0, g, T_0)
    Temp_2k  = T_of_m(m_2k , Delta_0, delta_V, 200e6,   g, T_0)

    # Domínio comum
    T_min = max(np.nanmin(Temp_amb), np.nanmin(Temp_2k))
    T_max = min(np.nanmax(Temp_amb), np.nanmax(Temp_2k))
    if not np.isfinite(T_min) or not np.isfinite(T_max) or T_max <= T_min:
        return None  # sinaliza inválido

    T_common = np.linspace(T_min, T_max, 1000)

    # Fração HS
    n_hs_amb = (m_amb + 1.0)/2.0
    n_hs_2k  = (m_2k  + 1.0)/2.0
    n_hs_amb_interp = np.interp(T_common, Temp_amb, n_hs_amb)
    n_hs_2k_interp  = np.interp(T_common, Temp_2k , n_hs_2k )

    # Entropias
    S_C_amb   = S_C_from_nhs(n_hs_amb_interp)
    S_C_2k    = S_C_from_nhs(n_hs_2k_interp)
    dS_conf   = -(S_C_2k - S_C_amb)

    Sspin     = S_S(T_common, 2.0, 2.0, 0.01)
    dS_spin   = -(n_hs_2k_interp*Sspin - n_hs_amb_interp*Sspin)

    # Óptica
    S_O_HS = np.array([S_O(3, T, FREQ_HS) for T in T_common])
    S_O_LS = np.array([S_O(3, T, FREQ_LS) for T in T_common])
    S_O_2k  = n_hs_2k_interp*S_O_HS + (1.0 - n_hs_2k_interp)*S_O_LS
    S_O_amb = n_hs_amb_interp*S_O_HS + (1.0 - n_hs_amb_interp)*S_O_LS
    dS_opt  = -(S_O_2k - S_O_amb)

    dS_total = dS_conf + dS_spin + dS_opt

    # Máximo
    idx_max = int(np.nanargmax(dS_total))
    max_val = float(dS_total[idx_max])
    max_T   = float(T_common[idx_max])

    return {
        "T_common": T_common,
        "dS_conf": dS_conf,
        "dS_spin": dS_spin,
        "dS_opt":  dS_opt,
        "dS_total": dS_total,
        "max_val": max_val,
        "max_T": max_T,
        "n_hs_amb": n_hs_amb_interp,
        "n_hs_2k": n_hs_2k_interp
    }

# ================================
# Função objetivo da otimização
# ================================
def objective_freevars(x, free_names, fixed_params):
    # Reconstrói parâmetros completos
    params = fixed_params.copy()
    for name, val in zip(free_names, x):
        params[name] = val

    curves = compute_curves(params["g"], params["Delta_0"], params["T_0"], params["delta_V"])
    if curves is None or not np.isfinite(curves["max_val"]):
        return 1e6  # penalização pesada

    # alvo (conforme seu código)
    target_max  = 70.0
    target_temp = 313.0
    err_max  = abs(curves["max_val"] - target_max)/target_max
    err_temp = abs(curves["max_T"]   - target_temp)/target_temp
    return 0.7*err_temp + 0.3*err_max

# ================================
# Helpers: separar livres/fixos/sweep e bounds/chutes
# ================================
def get_free_names_and_bounds(pc):
    names, bounds = [], []
    for k, spec in pc.items():
        if spec["mode"] == "optimize":
            names.append(k)
            bounds.append(spec["bounds"])
    return names, bounds

def random_initial_guess(bounds):
    return [np.random.uniform(low, high) for (low, high) in bounds]

def get_sweep_grid(pc):
    # Retorna lista de dicionários com {param: valor} para todos os parâmetros "sweep".
    sweep_keys = []
    sweep_vals = []
    for k, spec in pc.items():
        if spec["mode"] == "sweep":
            vals = spec.get("values", None)
            if vals is None or len(vals) == 0:
                raise ValueError(f'param "{k}" com mode="sweep" exige "values".')
            sweep_keys.append(k)
            sweep_vals.append(vals)
    if not sweep_keys:
        return [dict()]  # sem sweep → 1 combinação vazia
    combos = []
    for tup in product(*sweep_vals):
        combos.append({k:v for k, v in zip(sweep_keys, tup)})
    return combos

def build_fixed_params_for_combo(pc, combo):
    # Constrói dicionário de parâmetros completos, usando:
    # - valores do combo para "sweep"
    # - value para "fixed"
    # - placeholder (value) para "optimize", será sobrescrito pela otimização
    params = {}
    for k, spec in pc.items():
        if spec["mode"] == "sweep":
            params[k] = combo[k]
        elif spec["mode"] == "fixed":
            params[k] = spec["value"]
        elif spec["mode"] == "optimize":
            # placeholder provisório (não usado diretamente no modelo sem passar pela otimização)
            # mas é útil para imprimir ou default
            params[k] = spec.get("value", None)
        else:
            raise ValueError(f"mode inválido para {k}: {spec['mode']}")
    return params

# ================================
# Execução: varreduras + otimizações + plots
# ================================
def run_all():
    free_names, bounds = get_free_names_and_bounds(param_control)
    sweep_combos = get_sweep_grid(param_control)

    print(f"Parâmetros livres: {free_names}")
    print(f"Número de combinações (sweep): {len(sweep_combos)}")
    print(f"Testes por combinação: {N_TESTS}")

    for combo_idx, combo in enumerate(sweep_combos, start=1):
        base_params = build_fixed_params_for_combo(param_control, combo)

        for test_idx in range(1, N_TESTS+1):
            # Se houver parâmetros livres, otimiza; caso contrário, apenas plota com base_params
            if free_names:
                x0 = random_initial_guess(bounds)
                res = minimize(objective_freevars, x0,
                               args=(free_names, base_params),
                               bounds=bounds, method="L-BFGS-B")
                opt_params = base_params.copy()
                for name, val in zip(free_names, res.x):
                    opt_params[name] = float(val)
                fval = float(res.fun)
            else:
                opt_params = base_params.copy()
                fval = np.nan

            # Checagem: todos os 4 parâmetros devem existir no dicionário final
            for k in ["g", "Delta_0", "T_0", "delta_V"]:
                if k not in opt_params or opt_params[k] is None:
                    # se algum estava como "optimize" e não rodamos otimização (caso sem livres),
                    # usar o "value" configurado
                    opt_params[k] = param_control[k]["value"]

            # Curvas para o plot
            curves = compute_curves(opt_params["g"], opt_params["Delta_0"],
                                    opt_params["T_0"], opt_params["delta_V"])
            if curves is None:
                print(f"[AVISO] Combinação {combo_idx}/{len(sweep_combos)}, teste {test_idx}: intervalo T inválido. Pulando.")
                continue

            # Plot
            plt.figure(figsize=(12, 8))
            plt.xlim(250, 450)
            plt.scatter(delta_S_p2_rev_comp['T(K)'], delta_S_p2_rev_comp['DS(J/kgK)'],
                        color='black', label='Experimental')

            plt.plot(curves["T_common"], curves["dS_opt"],  label=r'$\Delta S_{\mathrm{opt}}$')
            plt.plot(curves["T_common"], curves["dS_spin"], label=r'$\Delta S_{\mathrm{spin}}$')
            plt.plot(curves["T_common"], curves["dS_conf"], label=r'$\Delta S_{\mathrm{config}}$')
            plt.plot(curves["T_common"], curves["dS_total"], label=r'$\Delta S_{\mathrm{total}} = \Delta S_{\mathrm{spin}} + \Delta S_{\mathrm{config}} + \Delta S_{\mathrm{opt}}$', linewidth=3)

            # Máximo
            plt.axvline(x=curves["max_T"], linestyle='--', alpha=0.7)
            plt.axhline(y=curves["max_val"], linestyle='--', alpha=0.7)

            # Título e labels
            g_ = opt_params["g"]; D0_ = opt_params["Delta_0"]; T0_ = opt_params["T_0"]; dV_ = opt_params["delta_V"]
            title = (f"Comb {combo_idx}/{len(sweep_combos)} | Teste {test_idx} | "
                     f"g={g_:.2f}, Δ₀={D0_:.2f} K, T₀={T0_:.2f} K, ΔV={dV_:.2e} m³")
            if np.isfinite(fval):
                title += f" | Obj={fval:.5f}"
            plt.title(title)
            plt.ylabel(r"$\Delta S$ (J/kg·K)")
            plt.xlabel("Temperature (K)")
            plt.legend()
            plt.grid(True, alpha=0.3)

            # Nome do arquivo
            combo_tag = "_".join([f"{k}_{combo[k]:.3e}" for k in combo.keys()]) if combo else "nosweep"
            fname = (f"comb{combo_idx:02d}_test{test_idx:02d}_"
                     f"g_{g_:.2f}_Delta0_{D0_:.2f}_T0_{T0_:.2f}_dV_{dV_:.3e}.png")
            if combo_tag != "nosweep":
                fname = f"{combo_tag}__" + fname
            filepath = os.path.join(OUTPUT_DIR, fname)
            plt.savefig(filepath, dpi=300, bbox_inches='tight')
            plt.close()

            print(f"[OK] Salvo: {filepath}")

# ================================
# Rodar tudo
# ================================
if __name__ == "__main__":
    run_all()
```
