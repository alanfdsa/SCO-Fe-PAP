```python
import re

input_file = "LS_modes.xyz"   # coloque o nome correto do seu arquivo aqui
output_file = "LS_modes_freq.txt"

frequencias = []

with open(input_file, "r") as f:
    for line in f:
        match = re.search(r"Mode\s+\d+\s*,\s*([\d\.]+)\s*1/cm", line)
        if match:
            freq = float(match.group(1))
            frequencias.append(freq)

# Salva em CSV
with open(output_file, "w") as f:
    f.write("Modo,Frequencia (cm-1)\n")
    for i, freq in enumerate(frequencias, start=1):
        f.write(f"{i},{freq:.4f}\n")

print(f"✔ Frequências extraídas: {len(frequencias)}")
print(f"Arquivo gerado: {output_file}")
```

```python
arquivo = "LS_modes.xyz"
num_atoms = 53

dados = []

with open(arquivo, "r") as f:
    linhas = f.readlines()

modo_atual = None
atom_count = 0

for linha in linhas:
    match_mode = re.search(r"Mode\s+(\d+)\s*,", linha)
    if match_mode:
        modo_atual = int(match_mode.group(1))
        atom_count = 0
        continue

    if modo_atual is not None:
        partes = linha.split()
        if len(partes) == 7:  # Elemento + 6 números
            atom_count += 1
            dx, dy, dz = map(float, partes[-3:])
            dados.append([modo_atual, atom_count, dx, dy, dz])

            if atom_count == num_atoms:
                modo_atual = None

df_disp = pd.DataFrame(dados, columns=["Mode", "AtomIndex", "dx", "dy", "dz"])
df_disp.to_csv("LS_modes_disp.txt", index=False)
```

```python
import numpy as np
df = pd.read_csv("HS_modes_disp.txt")

# Filtrar apenas o modo 1
modo1 = df[df["Mode"] == 1]

# Índices de interesse
selecionados = [53, 21, 22, 27, 28, 24, 29]

# Soma dos quadrados dos deslocamentos dos selecionados
sel = modo1[modo1["AtomIndex"].isin(selecionados)]
contrib_sel = np.sum(sel["dx"]**2 + sel["dy"]**2 + sel["dz"]**2)

# Soma total dos quadrados para todos os 53 átomos
contrib_total = np.sum(modo1["dx"]**2 + modo1["dy"]**2 + modo1["dz"]**2)

# Fração de participação dos átomos selecionados
fracao = contrib_sel / contrib_total

print(f"Soma selecionados = {contrib_sel:.6f}")
print(f"Soma total = {contrib_total:.6f}")
print(f"Fração participação = {fracao:.6f}")
```

```python
import pandas as pd
import numpy as np

# Carregar o arquivo
df = pd.read_csv("HS_modes_disp.txt")

# 1️⃣ Filtrar apenas o modo 1
df_mode1 = df[df["Mode"] == 1].copy()

# 2️⃣ Lista com símbolo atômico de cada índice
atom_symbols = [
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","N","N","O","O","O","N","N","O",
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","O","O","Fe"
]

# 3️⃣ Massas atômicas (u)
atomic_masses = {
    "H": 1.0079,
    "C": 12.011,
    "N": 14.007,
    "O": 15.999,
    "Fe": 55.845
}

# Adicionar coluna com massas
df_mode1["Mass"] = [atomic_masses[sym] for sym in atom_symbols]

# 4️⃣ Calcular m * (dx² + dy² + dz²)
df_mode1["MassWeightedDisp2"] = df_mode1["Mass"] * (
    df_mode1["dx"]**2 + df_mode1["dy"]**2 + df_mode1["dz"]**2
)

# Átomos específicos do cluster (índices: Fe=53, N1=21, N2=22, N3=27, N4=28, O1=24, O2=29)
cluster_atoms = [53, 21, 22, 27, 28, 24, 29]

# 5️⃣ Somar contribuições
cluster_sum = df_mode1[df_mode1["AtomIndex"].isin(cluster_atoms)]["MassWeightedDisp2"].sum()
total_sum = df_mode1["MassWeightedDisp2"].sum()

fraction = cluster_sum / total_sum

print("Contribuição do cluster (com massas):", cluster_sum)
print("Contribuição total:", total_sum)
print("Fraçao cluster/total:", fraction)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# HS
df_hs = pd.read_csv("HS_modes_disp.txt")
atom_symbols_hs = [
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","N","N","O","O","O","N","N","O",
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","O","O","Fe"
]
atomic_masses = {
    "H": 1.0079,
    "C": 12.011,
    "N": 14.007,
    "O": 15.999,
    "Fe": 55.845
}
cluster_atoms_hs = [53, 21, 22, 27, 28, 24, 29]
df_hs["Mass"] = [atomic_masses[atom_symbols_hs[i-1]] for i in df_hs["AtomIndex"]]
df_hs["MassWeightedDisp2"] = df_hs["Mass"] * (df_hs["dx"]**2 + df_hs["dy"]**2 + df_hs["dz"]**2)
modes_hs = sorted(df_hs["Mode"].unique())
fractions_hs = []
for mode in modes_hs:
    df_mode = df_hs[df_hs["Mode"] == mode]
    cluster_sum = df_mode[df_mode["AtomIndex"].isin(cluster_atoms_hs)]["MassWeightedDisp2"].sum()
    total_sum = df_mode["MassWeightedDisp2"].sum()
    frac = cluster_sum / total_sum if total_sum > 0 else 0.0
    fractions_hs.append(frac)

# LS
df_ls = pd.read_csv("LS_modes_disp.txt")
atom_symbols_ls = [
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","N","N","O","O","O",
    "C","H","C","H","C","H","C","H","C","C","H","C","C","C","H",
    "C","H","C","C","H","N","N","N","O","O","O","Fe"
]
cluster_atoms_ls = [53, 21, 22, 47, 48, 24, 50]
df_ls["Mass"] = [atomic_masses[atom_symbols_ls[i-1]] for i in df_ls["AtomIndex"]]
df_ls["MassWeightedDisp2"] = df_ls["Mass"] * (df_ls["dx"]**2 + df_ls["dy"]**2 + df_ls["dz"]**2)
modes_ls = sorted(df_ls["Mode"].unique())
fractions_ls = []
for mode in modes_ls:
    df_mode = df_ls[df_ls["Mode"] == mode]
    cluster_sum = df_mode[df_mode["AtomIndex"].isin(cluster_atoms_ls)]["MassWeightedDisp2"].sum()
    total_sum = df_mode["MassWeightedDisp2"].sum()
    frac = cluster_sum / total_sum if total_sum > 0 else 0.0
    fractions_ls.append(frac)

# Gráfico combinado
plt.figure()
plt.plot(modes_hs, fractions_hs, marker="o", color='black', label='HS')
plt.plot(modes_ls, fractions_ls, marker="o", color='red', label='LS')
plt.xlabel("Número do Modo Normal")
plt.ylabel("Fração cluster/total (com massas)")
plt.title("Contribuição do cluster em cada modo normal")
plt.legend()
plt.grid(True)
plt.show()
```
