```python
def renomear_index_por_atom(df, lista_atomica):
    """
    df: DataFrame final já criado (159 linhas)
    lista_atomica: lista de strings com linhas contendo símbolo, massa e coordenadas
    """
    # Extrair apenas o símbolo atômico de cada linha
    simbolos = [linha.split()[0] for linha in lista_atomica]

    novo_index = []
    contadores = {}  # exemplo: {"C":1, "H":1, "O":1}

    for simbolo in simbolos:
        if simbolo not in contadores:
            contadores[simbolo] = 1

        nome = f"{simbolo}{contadores[simbolo]}"

        # Cada átomo corresponde a 3 linhas do DataFrame
        novo_index.extend([nome, nome, nome])

        contadores[simbolo] += 1

    # Aplicar ao DataFrame
    df.index = novo_index
    return df
```

```python
arquivo = "ls/v2/vib_mode.txt"

with open(arquivo, "r") as f:
    linhas = f.readlines()

# --------------------
# 1) Parte inicial (159 linhas)
# --------------------

linhas_159 = linhas[:159]

header = re.split(r"\s+", linhas[0].strip())
header = [h for h in header if h != ""]

df = pd.read_csv(
    StringIO("".join(linhas_159)),
    delim_whitespace=True,
    names=header,
    skiprows=1
)

# --------------------
# 2) Blocos seguintes
# --------------------

blocos = []
i = 159

num_regex = r"^\s*\d+(\s+\d+)*\s*$"

while i < len(linhas):

    # Linha com números das colunas
    col_header_line = linhas[i].strip()
    novas_cols = re.split(r"\s+", col_header_line)
    novas_cols = [c for c in novas_cols if c != ""]
    i += 1

    bloco_linhas = []

    # Acumular linhas do bloco
    while i < len(linhas) and not re.match(num_regex, linhas[i]):
        bloco_linhas.append(linhas[i])
        i += 1

    # Se não houver linhas, pula
    if len(bloco_linhas) == 0:
        continue

    # ----------------------------
    # Detectar número real de colunas no bloco
    # ----------------------------
    primeira_linha = bloco_linhas[0].strip()
    ncols = len(re.split(r"\s+", primeira_linha))

    # criar nomes para todas as colunas
    # primeira é índice, outras são novas_cols (até onde existirem)
    nomes = ["idx"] + novas_cols + [f"extra_{k}" for k in range(len(novas_cols)+1, ncols)]

    df_bloco = pd.read_csv(
        StringIO("".join(bloco_linhas)),
        delim_whitespace=True,
        names=nomes
    )

    df_bloco = df_bloco.drop(columns=["idx"])
    blocos.append(df_bloco)

# --------------------
# 3) Concatenar
# --------------------

for b in blocos:
    df = pd.concat([df, b], axis=1)
```

```python
df = renomear_index_por_atom(df, lista_atomica)
normal_modes_ls = df
```

```python
# -------------------------
# 1) Monta a matriz (linhas = HS i, colunas = LS j)
# -------------------------
ini = 0
#ini = 6
lista_array_fin = []
for i in range(ini,159):
    iesimo_hs_nm = normal_modes_hs[str(i)]   # HS
    row = []
    for j in range(ini,159):
        jesimo_ls_nm = normal_modes_ls[str(j)]  # LS
        prod_intern_all = (iesimo_hs_nm * jesimo_ls_nm).sum()
        #row.append(abs(prod_intern_all))
        row.append((prod_intern_all))
    lista_array_fin.append(row)

prod_intern_all = pd.DataFrame(lista_array_fin)   # shape (HS, LS)
mat_all = prod_intern_all.to_numpy()
nrows_all, ncols_all = mat_all.shape
```

```python
import numpy as np

def svd_manual_block(
    mat_all,
    hs_modes,
    ls_modes,
    valores_modos_hs=None,
    valores_modos_ls=None,
    top_n=None,
    remove_ls_below_hs=False,
    calculate_effective_frequency=True
):
    """
    mat_all[HS, LS] = overlap com sinal

    hs_modes e ls_modes são índices Python.

    valores_modos_hs[row] = frequência do modo HS row
    valores_modos_ls[col] = frequência do modo LS col

    top_n:
        Se definido, mantém apenas as top_n maiores contribuições
        em |overlap|² dentro do range escolhido.

    remove_ls_below_hs:
        Se True, remove modos LS cuja frequência seja menor
        que a frequência do modo HS analisado.

    calculate_effective_frequency:
        Calcula frequência LS efetiva ponderada por |overlap|².
    """

    hs_idx = hs_modes
    ls_idx_original = ls_modes

    print("\nPesos individuais |overlap|²:")

    for h_mode, h_idx in zip(hs_modes, hs_idx):

        ls_idx = list(ls_idx_original)

        # -------------------------------------------------
        # Remove LS com frequência menor que HS
        # -------------------------------------------------
        if remove_ls_below_hs:

            if valores_modos_hs is None or valores_modos_ls is None:
                raise ValueError(
                    "Para usar remove_ls_below_hs=True, "
                    "é necessário fornecer valores_modos_hs e valores_modos_ls."
                )

            hs_freq = float(valores_modos_hs[h_idx])

            ls_idx = [
                l_idx for l_idx in ls_idx
                if float(valores_modos_ls[l_idx]) >= hs_freq
            ]

        # -------------------------------------------------
        # Seleciona top_n maiores contribuições
        # -------------------------------------------------
        if top_n is not None:

            weights_all = np.abs(mat_all[h_idx, ls_idx])**2

            top_indices_local = np.argsort(weights_all)[::-1][:top_n]

            ls_idx = [ls_idx[k] for k in top_indices_local]

            ls_idx = sorted(ls_idx)

        # -------------------------------------------------
        # SVD
        # -------------------------------------------------
        block = mat_all[np.ix_([h_idx], ls_idx)]

        U, S, Vt = np.linalg.svd(block)

        print("\n" + "="*60)

        print("Bloco analisado (HS x LS):")
        print(block)

        # -------------------------------------------------
        # HS
        # -------------------------------------------------
        print("\nModos HS analisados:")

        if valores_modos_hs is not None:
            hs_freq = float(valores_modos_hs[h_idx])
            print(f"  HS {h_mode}: freq = {hs_freq:.3f}")
        else:
            hs_freq = None
            print(f"  HS {h_mode}")

        # -------------------------------------------------
        # LS
        # -------------------------------------------------
        print("\nModos LS considerados:")

        for l_mode in ls_idx:

            if valores_modos_ls is not None:
                ls_freq = float(valores_modos_ls[l_mode])
                print(f"  LS {l_mode}: freq = {ls_freq:.3f}")
            else:
                print(f"  LS {l_mode}")

        # -------------------------------------------------
        # Singular values
        # -------------------------------------------------
        print("\nValores singulares:")

        for k, sigma in enumerate(S, start=1):
            print(f"  sigma_{k} = {sigma:.5f}")

        # -------------------------------------------------
        # Interpretação
        # -------------------------------------------------
        print("\nInterpretação do match:")

        if hs_freq is not None:
            hs_label = f"HS {h_mode} ({hs_freq:.3f})"
        else:
            hs_label = f"HS {h_mode}"

        print(
            f"O modo normal {hs_label} é uma combinação linear "
            f"dos modos LS {', '.join(map(str, ls_idx))}."
        )

        # -------------------------------------------------
        # Pesos
        # -------------------------------------------------
        weights = np.abs(mat_all[h_idx, ls_idx])**2

        total = np.sum(weights)

        order = np.argsort(weights)[::-1]

        print("\nPesos individuais |overlap|²:")

        if hs_freq is not None:
            print(f"\nHS {h_mode} | freq = {hs_freq:.3f}:")
        else:
            print(f"\nHS {h_mode}:")

        for idx in order:

            l_mode = ls_idx[idx]

            w = weights[idx]

            overlap = mat_all[h_idx, l_mode]

            if valores_modos_ls is not None:

                ls_freq = float(valores_modos_ls[l_mode])

                print(
                    f"  LS {l_mode} | freq = {ls_freq:.3f} | "
                    f"overlap = {overlap:.6f} | contribuição = {w:.6f}"
                )

            else:

                print(
                    f"  LS {l_mode} | overlap = {overlap:.6f} | "
                    f"contribuição = {w:.6f}"
                )

        print(f"  soma no subespaço LS = {total:.6f}")

        # -------------------------------------------------
        # Frequência efetiva
        # -------------------------------------------------
        if (
            calculate_effective_frequency
            and valores_modos_ls is not None
        ):

            ls_freqs = np.array([
                float(valores_modos_ls[l_mode])
                for l_mode in ls_idx
            ])

            if total > 0:

                f_ls_eff = np.sum(weights * ls_freqs) / total

                print("\nFrequência efetiva LS:")

                print(f"  f_LS_eff = {f_ls_eff:.3f}")

                if hs_freq is not None:

                    delta_f = f_ls_eff - hs_freq

                    print(f"  f_HS     = {hs_freq:.3f}")

                    print(f"  Δf       = {delta_f:.3f}")
```
